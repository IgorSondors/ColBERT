```python
def clear_df(df):
    # df = df[df["model_id"] != 0]
    df.drop_duplicates(subset="name", keep="first", inplace=True)
    return df.reset_index(drop=True)

def choose_true_matches_models(df, df_models):
    keys = list(df_models['model_id'])
    values = list(df_models['full_name'])
    id_to_name = dict(zip(keys, values))

    where_is_model_id = []
    for i in set(list(df['model_id'])):
        if not i in keys:
            where_is_model_id.append(i)
    for i in where_is_model_id: # среди офферов существуют model_id, которых нет среди моделей, имеющихся у меня на данный момент
        df = df[df['model_id']!=i]

    model_gt = [id_to_name[i] for i in df['model_id']]
    df['true_match'] = model_gt

    return df

df_raw_unique = clear_df(df_raw)
# df_offers = choose_true_matches_models(triplets_offer_model, df_models)

clms = ['attrs',
        'price',
        'description',
        'external_category',
        'external_brand',
        'external_type'
        ]

df_raw_unique = df_raw_unique.drop(columns=clms)
df_raw_unique
```

```python
import os

for model_id_pred in set(df_id0_labeled.model_id_pred_1):
    model = list(df_models13[df_models13.model_id == model_id_pred]['full_name'])[0]
    cat_id = list(df_models13[df_models13.model_id == model_id_pred]['category_id'])[0]
    cat = list(df_models13[df_models13.model_id == model_id_pred]['category_name'])[0]
    dst_fld = f"/home/sondors/Documents/price/ColBERT/050224_Timofey/clusters/{cat_id}"
    os.makedirs(dst_fld, exist_ok=True)

    df_by_model_id = df_id0_labeled[df_id0_labeled.model_id_pred_1 == model_id_pred]

    with open(f"{dst_fld}/{len(df_by_model_id)}_{model_id_pred}.txt", "a") as txt:
        txt.write(f"category_id: {cat_id}\ncategory_name: {cat}\nmodel_id_pred_1: {model_id_pred}\nmodel: {model}\n\n")
        for index, row in df_by_model_id.iterrows():
            txt.write(f"\n{row['name']} --> {row['similarity_1']}")
```
