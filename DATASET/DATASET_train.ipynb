```python
import pandas as pd
import random
import os

scr_pth = "/home/sondors.igor/documents/ColBERT/DATASET/train_triplets.parquet"
dst_fld = "/home/sondors.igor/documents/ColBERT/DATASET/train_triplets/15mln"
os.makedirs(dst_fld, exist_ok=True)

dst_name = "v0"
clms = ['name', 'false_match', 'true_match']
df_triplets = pd.read_parquet(scr_pth)[clms]
df_triplets
```

```python
def fn(x):
    return x.strip().replace("\n", " ").replace("\t", " ")
for c in clms:
    df_triplets[c] = df_triplets[c].apply(fn)
```

```python
"""
The file documents_train.tsv contains the training documents and has the following format:
0 text0
1 text1
2 text2
"""

documents = set(list(df_triplets['true_match'])+list(df_triplets['false_match']))
id = [i for i in range(len(documents))]
id_to_document = dict(zip(id, documents))
document_to_id = {y: x for x, y in id_to_document.items()}

documents_train = pd.DataFrame()
documents_train["id"] = id
documents_train["documents"] = list(documents)

# id_to_document
# document_to_id
documents_train.to_csv(f'{dst_fld}/documents_train_{dst_name}.tsv', sep='\t', header=False, index=False)

documents_train
```

```python
"""
The file queries.tsv contains the queries and has the following format:
0 query0
1 query1
2 query2
"""

queries = list(df_triplets.drop_duplicates(subset=['name'], keep="first")['name'])
id = [i for i in range(len(queries))]
id_to_query = dict(zip(id, queries))
query_to_id = {y: x for x, y in id_to_query.items()}

queries_train = pd.DataFrame()
queries_train["id"] = id
queries_train["queries"] = queries

# id_to_query
# query_to_id
queries_train.to_csv(f'{dst_fld}/queries_train_{dst_name}.tsv', sep='\t', header=False, index=False)

queries_train
```

```python
"""
...
The file triples.json contains the training triples and has the following format:
[0, 12692, 48199]
[0, 12693, 17664]
[1, 23783, 21169]
[1, 3796, 17104]
[1, 47897, 30084]
[2, 52241, 54085]
...
The first component is the id of the query, the second one the id of a relevant 
document and the third one the id of an irrelevant document.
"""

with open(f'{dst_fld}/triples_X1_{dst_name}.json', "a") as json:
    for index, row in df_triplets.iterrows():
        query_id = query_to_id[row['name']]
        true_match_id = document_to_id[row['true_match']]
        false_match_id = document_to_id[row['false_match']]

        json.write(f"[{query_id}, {true_match_id}, {false_match_id}]\n")
```

```python
"""
Shuffle json and save
"""

def shuffle_wrt_file(output_file_path, lines):
    random.shuffle(lines)
    with open(output_file_path, 'w', encoding='utf-8') as output_file:
        output_file.writelines(lines)

# Открываем исходный файл для чтения
with open(f'{dst_fld}/triples_X1_{dst_name}.json', 'r', encoding='utf-8') as input_file:
    # Читаем строки из исходного файла
    lines = input_file.readlines()

output_file_path = f'{dst_fld}/triples_X1_{dst_name}_shuffle.json'

shuffle_wrt_file(output_file_path, lines)
print(f'Файл {output_file_path} успешно создан с перемешанным порядком строк.')
```
