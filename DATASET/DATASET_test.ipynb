```python
def all_categories_together(pth_triplets, pth_models, fld_dst="/mnt/vdb1/Datasets/ColBERT/18_categories/metrics_data"):
    query, query_id = split(pth_triplets, col="name")
    query.to_csv(f'{fld_dst}/offers.tsv', sep='\t', header=False, index=False)
    query_id.to_csv(f'{fld_dst}/offers_id.tsv', sep='\t', header=False, index=False)
    print(query,"\n" ,query_id)

    
    document, document_id = split(pth_models, col="full_name")
    document.to_csv(f'{fld_dst}/models.tsv', sep='\t', header=False, index=False)
    document_id.to_csv(f'{fld_dst}/models_id.tsv', sep='\t', header=False, index=False)
    print(document,"\n" ,document_id)
    
def split(pth, col="name"):
    df_my = pd.read_csv(pth, sep=';')
    df1 = pd.DataFrame()
    df1["id"], df1[col] = [i for i in range(len(df_my))], df_my[col]
    
    df2 = pd.DataFrame()
    df2["id"], df2["model_id"] = [i for i in range(len(df_my))], df_my['model_id']
    return df1, df2

pth_triplets = "/mnt/vdb1/Datasets/ColBERT/18_categories/metrics_data/triplets_test_18_categories.csv"
pth_models = "/mnt/vdb1/Datasets/ColBERT/18_categories/metrics_data/models_18_categories.csv"
# all_categories_together(pth_triplets, pth_models) # старый тестовый датасет на поиск матча среди всех категорий вперемешку

```
