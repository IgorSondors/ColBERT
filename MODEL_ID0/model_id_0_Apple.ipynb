```python
# Создание массива значений x
x_values = np.linspace(0, 1, 100)  # 100 значений от 0 до 1

# Создание списка для хранения процентных значений
percentages = []

# Вычисление доли строк для каждого значения x
for x in x_values:
    filtered_df = merged_df[merged_df['cosine_sims_1'] >= x]
    percentage = len(filtered_df) / len(df_offers) * 100
    percentages.append(percentage)

# Построение графика
plt.plot(x_values, percentages, color='skyblue')
plt.title('Процент строк, удовлетворяющих условию')
plt.xlabel('cosine_sims_1')
plt.ylabel('% профита')
plt.xticks(np.arange(0, 1.1, 0.1))  # Задание цены деления по оси X
plt.yticks(np.arange(0, 12, 0.5))   # Задание цены деления по оси Y
plt.grid(True)
plt.show()
```
