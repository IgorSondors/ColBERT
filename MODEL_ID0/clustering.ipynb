```python
from sklearn.cluster import DBSCAN
import itertools
import operator

def most_common(L):
    # get an iterable of (item, iterable) pairs
    SL = sorted((x, i) for i, x in enumerate(L))
    # print 'SL:', SL
    groups = itertools.groupby(SL, key=operator.itemgetter(0))
    # auxiliary function to get "quality" for an item
    def _auxfun(g):
        item, iterable = g
        count = 0
        min_index = len(L)
        for _, where in iterable:
            count += 1
            min_index = min(min_index, where)
            # print 'item %r, count %r, minind %r' % (item, count, min_index)
        return count, -min_index
    # pick the highest-count/earliest item
    return max(groups, key=_auxfun)[0]

def cluster_accuracy(cluster):
    cluster_gt = most_common(cluster)
    n_all = len(cluster)
    n_correct = cluster.count(cluster_gt)
    return n_correct / n_all

def labels_to_clusters(db_labels, model_id):
    clusters = [[] for i in range(len(set(db_labels)) - 1)]
    for i in range(len(model_id)):
        cluster_id = db_labels[i]
        if cluster_id == -1:
            continue
        clusters[cluster_id].append(model_id[i])
    return clusters    

def acc_mean(db_labels, model_id):
    clusters = labels_to_clusters(db_labels, model_id)

    cluster_n = -1
    accuracy_mean = 0
    for cluster in clusters:
        cluster_n += 1
        # if len(cluster) < 20:
        #     continue
        accuracy = cluster_accuracy(cluster)
        accuracy_mean += accuracy/len(clusters)
        # print(f"cluster {cluster_n}:\naccuracy = {round(100 * accuracy, 1)}%, len = {len(cluster)}\n")
    return accuracy_mean

def find_threshold(data, model_id, metric="precomputed"):
    epsilon_list = []
    accuracy_list = []
    for i in range(1000):
        print(i)
        epsilon = 0.000001+i/251
        db = DBSCAN(eps=epsilon, min_samples=4, metric=metric).fit(data)
        db_labels = db.labels_
        accuracy = acc_mean(db_labels, model_id)
        print(f"epsilon = {epsilon}:\naccuracy = {round(100 * accuracy, 1)}%\n")
        epsilon_list.append(epsilon)
        accuracy_list.append(accuracy)
    return epsilon_list, accuracy_list
```
