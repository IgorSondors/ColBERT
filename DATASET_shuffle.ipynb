```python
import random

# Открываем исходный файл для чтения
input_file_path = '/mnt/vdb1/Datasets/ColBERT/18_categories/train/triples.json'
with open(input_file_path, 'r', encoding='utf-8') as input_file:
    # Читаем строки из исходного файла
    lines = input_file.readlines()

lines
```

```python
def shuffle_wrt_file(output_file_path, lines):
    # Перемешиваем порядок строк
    random.shuffle(lines)

    # Открываем новый файл для записи
    with open(output_file_path, 'a', encoding='utf-8') as output_file:
        # Записываем перемешанные строки в новый файл
        output_file.writelines(lines)

multiplier = 12
output_file_path = f'/mnt/vdb1/Datasets/ColBERT/18_categories/train/triples_X{multiplier}_shuffle.json'
for i in range(multiplier):
    shuffle_wrt_file(output_file_path, lines)

print(f'Файл {output_file_path} успешно создан с перемешанным порядком строк.')
```
